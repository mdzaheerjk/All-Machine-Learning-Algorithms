### **Example Dataset**

Suppose we have 6 points in 2D:

| Point | x  | y |
| ----- | -- | - |
| A     | 1  | 2 |
| B     | 1  | 4 |
| C     | 1  | 0 |
| D     | 10 | 2 |
| E     | 10 | 4 |
| F     | 10 | 0 |

We want to cluster them into **2 clusters** ($K=2$).

---

### **Step 1: Initialize Centroids**

Randomly pick 2 points as centroids. Let’s pick:

* Centroid 1: A (1, 2)
* Centroid 2: D (10, 2)

---

### **Step 2: Assign Points to Nearest Centroid**

We calculate **Euclidean distance**:

* Distance formula:

$$
\text{Distance} = \sqrt{(x_2-x_1)^2 + (y_2-y_1)^2}
$$

| Point | Distance to Centroid1 (1,2) | Distance to Centroid2 (10,2) | Cluster |
| ----- | --------------------------- | ---------------------------- | ------- |
| A     | 0                           | √(9²+0²)=9                   | 1       |
| B     | √(0²+2²)=2                  | √(9²+2²)=√85 ≈ 9.22          | 1       |
| C     | √(0²+2²)=2                  | √(9²+2²)=√85 ≈ 9.22          | 1       |
| D     | 9                           | 0                            | 2       |
| E     | √(9²+2²)=√85 ≈ 9.22         | √(0²+2²)=2                   | 2       |
| F     | √(9²+2²)=√85 ≈ 9.22         | √(0²+2²)=2                   | 2       |

**Clusters after first assignment:**

* Cluster 1: A, B, C
* Cluster 2: D, E, F

---

### **Step 3: Update Centroids**

Calculate mean of each cluster:

* **Cluster 1 centroid:**

$$
x = (1+1+1)/3 = 1, \quad y = (2+4+0)/3 = 2
$$

New centroid = (1, 2) → same as before

* **Cluster 2 centroid:**

$$
x = (10+10+10)/3 = 10, \quad y = (2+4+0)/3 = 2
$$

New centroid = (10, 2) → same as before

---

### **Step 4: Convergence**

* Centroids did not change → algorithm **converged**.

---

### **Final Clusters**

* **Cluster 1:** A(1,2), B(1,4), C(1,0)
* **Cluster 2:** D(10,2), E(10,4), F(10,0)

# How to find K


### **Finding K using WCSS (Elbow Method)**
WCSS (Within-Cluster Sum of Squares)

1. **Try different K values**

   * Run K-means for K = 1, 2, 3, … up to maybe 10.

2. **Calculate WCSS for each K**

   * WCSS is **how far points are from their cluster center**.
   * Smaller WCSS = points are closer to their center.

3. **Make a plot**

   * X-axis: K (number of clusters)
   * Y-axis: WCSS

4. **Look for the “elbow”**

   * WCSS gets smaller as K increases, but at some point, it **stops decreasing quickly**.
   * That point looks like an **elbow in the curve**.

5. **Pick K at the elbow**

   * That K is usually the best number of clusters to use.

# K Means ++ Initialization

### 🔹 Problem with normal K-means

* K-means starts by picking **random cluster centers** (centroids).
* If the starting points are bad (too close together), the algorithm may:

  * Give poor clusters, or
  * Take longer to adjust.

---

### 🔹 What K-means++ does

It’s just a **smarter way to pick the starting points**.

* Instead of choosing centroids completely randomly, it **spreads them out**.

---

### 🔹 Steps (easy version)

1. Pick the **first centroid randomly** from your data.
2. Look at all other points → find their **distance to the nearest chosen centroid**.
3. The farther a point is, the **higher its chance** to become the next centroid.
4. Repeat until K centroids are chosen.
5. Then run normal K-means.

---

### 🔹 Why this helps

* Centroids start in **different areas of the data**.
* Clusters are usually **better**.
* Algorithm becomes **faster** and more reliable.

---

💡 Think of it like this:
If you’re planting **K trees** in a garden, you wouldn’t put all of them in one corner 🌳🌳🌳.
Instead, you’d spread them out.
K-means++ does the same thing for cluster centers.
