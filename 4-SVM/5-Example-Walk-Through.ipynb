## 🔹 Example: Points in a Circle

### 1. Dataset (2D: x, y)

Imagine we collect points:

* **Red class** = points **inside a circle** around the origin.
* **Blue class** = points **outside that circle**.

Sample points:

| Point | x    | y    | Class |
| ----- | ---- | ---- | ----- |
| A     | 0.1  | 0.2  | Red   |
| B     | 0.3  | 0.4  | Red   |
| C     | 0.2  | 0.1  | Red   |
| D     | 0.9  | 0.8  | Blue  |
| E     | 0.7  | 0.9  | Blue  |
| F     | 0.85 | 0.95 | Blue  |

---

### 2. Can we separate in 2D?

If you plot these on an **x-y plane**:

* Red points are clustered near the **center (0,0)**.
* Blue points surround them.

👉 No straight line can divide "inside the circle" vs "outside the circle."
So SVM with a **linear kernel** fails.

---

### 3. Add a higher dimension (z = x² + y²)

Now, let’s transform the data:

$$
z = x^2 + y^2
$$

| Point | x    | y    | z = x² + y² | Class |
| ----- | ---- | ---- | ----------- | ----- |
| A     | 0.1  | 0.2  | 0.05        | Red   |
| B     | 0.3  | 0.4  | 0.25        | Red   |
| C     | 0.2  | 0.1  | 0.05        | Red   |
| D     | 0.9  | 0.8  | 1.45        | Blue  |
| E     | 0.7  | 0.9  | 1.30        | Blue  |
| F     | 0.85 | 0.95 | 1.62        | Blue  |

---

### 4. Now in 3D (x, y, z)

* Red points have **small z values** (close to 0).
* Blue points have **large z values**.

👉 A simple flat plane like **z = 0.5** can now separate Red vs Blue.

---

### 5. Why is this important?

* In 2D: impossible with a straight line.
* In 3D: easy with a flat surface (hyperplane).
* **This is exactly what kernels do in SVM** — they map data into higher dimensions automatically, so separation becomes possible.

---

📌 **In short:**

* **Without kernel**: the SVM says “I can’t separate this circle data with a line.”
* **With kernel (like polynomial or RBF)**: the SVM “lifts” the data into a higher space and says “Oh! Now a straight line works perfectly.”
