**Isolation Forest** is a method for finding unusual points (anomalies) in your data. It works differently from many other methods because it doesn’t try to model “normal” behavior directly. Instead, it tries to **isolate each point** by randomly splitting the data.

Here’s the intuition:

* Anomalies are **few and different**, so they get isolated quickly with fewer splits.
* Normal points are **many and similar**, so they need more splits to be separated.

The algorithm builds a “forest” of many random trees, and then:

* Points that are isolated quickly get a **high anomaly score**.
* Points that take many splits to isolate get a **low anomaly score**.

In short: Isolation Forest is like a game of “divide and conquer” — the easier it is to separate a point from the rest, the more likely it’s an anomaly.
