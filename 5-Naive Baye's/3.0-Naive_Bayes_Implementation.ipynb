# Applying Multinomial Naive Bayes on the Iris dataset

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
mnb.fit(X_train,y_train)
y_predmnb=mnb.predict(X_test)
print(confusion_matrix(y_predmnb,y_test))
print(accuracy_score(y_predmnb,y_test))
print(classification_report(y_predmnb,y_test))
```

### MultinomialNB is not suitable for continuous data like the Iris dataset. It is designed for discrete data, such as word counts in text classification. Using MultinomialNB on continuous data can lead to poor performance, as seen in the results above. For continuous data, GaussianNB is the appropriate choice.

# Applying Bernouli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
bnb=BernoulliNB()
bnb.fit(X_train,y_train)
y_predbnb=bnb.predict(X_test)
print(confusion_matrix(y_predbnb,y_test))
print(accuracy_score(y_predbnb,y_test))
print(classification_report(y_predbnb,y_test))
```

### BernoulliNB is used for binary/boolean features , since our data is continuous it is not performing well.
