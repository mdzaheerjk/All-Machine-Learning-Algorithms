## üå≥ Hierarchical Clustering

Hierarchical clustering is a way to group data into clusters by **building a tree-like structure** (hierarchy).
Unlike K-means, you **don‚Äôt have to choose K in advance**.

---

### üîπ Two main types

1. **Agglomerative (bottom-up)** ‚Äì Start with every point as its **own cluster**, then keep **joining the closest clusters** until everything becomes one big cluster.
2. **Divisive (top-down)** ‚Äì Start with **all points in one cluster**, then keep **splitting into smaller clusters**.

(Mostly, people use **agglomerative**.)

---

### üîπ How it works (Agglomerative Example)

1. Start: each point is its **own cluster**.
   (If you have 5 points ‚Üí 5 clusters).
2. Find the **two closest clusters** and merge them.
3. Repeat step 2 until all points end up in **one big cluster**.
4. The process is shown in a **tree diagram** called a *dendrogram*.

---

### üîπ Why it‚Äôs useful

* You don‚Äôt need to say ‚ÄúI want 3 clusters‚Äù at the start.
* You can **look at the dendrogram** and decide how many clusters you want by ‚Äúcutting‚Äù the tree at a certain height.

---

### üîπ Simple Example

Imagine you have 6 students based on test scores.

* First, each student is their own group.
* Then the two students with the most similar scores are grouped.
* Slowly, groups keep merging until finally, all students are in one big group.
* You can ‚Äúcut‚Äù the tree at some level to say, ‚ÄúOkay, I want 2 groups of students.‚Äù

---

üí° Think of it like **family trees**:

* At the bottom = individual people (points).
* Higher up = families (clusters).
* At the top = one big family tree (all data).

### üîπ First Image ‚Äì Types of Hierarchical Clustering

* **Agglomerative (Top part of image)**

  * Start with every point as its **own small group**.
  * Step by step, merge the closest groups together.
  * Keep merging until all points become one big group.
  * Example in the image: small circles ‚Üí merge into bigger and bigger circles.

* **Divisive (Bottom part of image)**

  * Start with **all points together in one big group**.
  * Step by step, split them into smaller groups.
  * Keep splitting until each point stands alone.
  * Example in the image: one big oval ‚Üí splits into smaller ovals ‚Üí then into even smaller ones.

üëâ So, **Agglomerative = build up from small to big**,
**Divisive = break down from big to small**.

---

### üîπ Second Image ‚Äì Dendrogram

A **dendrogram** is just a **tree-like diagram** that shows how clusters form.

* **Bottom row (a, b, c, ‚Ä¶ n):** each item starts as its own point/cluster.
* **Branches joining together:** show which points or groups got merged together.
* **Height of the branch:** shows how far apart the groups were when they merged (closer = merge earlier, farther = merge later).
* **Red dashed line:** If you ‚Äúcut‚Äù the tree here, you get clusters.

  * In this example, cutting the line gives **4 clusters** (orange, green, blue, yellow).

üëâ So, dendrograms let you **see the whole process** and then **choose how many clusters** you want.

---

üí° In short:

* The **first image** shows the **two ways of building clusters** (bottom-up vs. top-down).
* The **second image** (dendrogram) shows the **merging process step by step** and helps you decide the number of clusters.

# Example

### üéì Example: Students‚Äô Exam Scores

We have 5 students with their test scores (out of 100):

* A = 10
* B = 15
* C = 17
* D = 90
* E = 92

Clearly, A, B, C are close in scores, and D, E are close.

---

### üîπ Step 1: Start (Agglomerative way)

* Each student is their **own cluster**: {A}, {B}, {C}, {D}, {E}

---

### üîπ Step 2: Merge the closest

* A and B are closest (10 & 15) ‚Üí merge into {A,B}
* Then add C (17), because it‚Äôs close ‚Üí {A,B,C}

---

### üîπ Step 3: Merge the other group

* D and E are close (90 & 92) ‚Üí merge into {D,E}

---

### üîπ Step 4: Final merge

* Now we have 2 big groups: {A,B,C} and {D,E}
* Finally, they merge into one big group {A,B,C,D,E}

---

### üîπ How this looks in a dendrogram

* At the bottom: A, B, C, D, E (each on its own branch).
* First branches join A and B, then add C.
* Another branch joins D and E.
* At the top, {A,B,C} merges with {D,E}.

If you **cut the dendrogram in the middle**, you get **2 clusters**:

* Cluster 1 ‚Üí {A, B, C} (low scores)
* Cluster 2 ‚Üí {D, E} (high scores)

---

üí° **Takeaway:**
Hierarchical clustering lets you see the whole ‚Äústory‚Äù of how clusters form. The dendrogram is just a tree that tells that story, and you decide how many groups to keep by cutting it at a certain level.
