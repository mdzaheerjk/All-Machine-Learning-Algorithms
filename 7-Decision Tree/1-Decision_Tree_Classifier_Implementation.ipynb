```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
##independent features
X=pd.DataFrame(iris['data'],columns=['sepal length','sepal width','petal length','petal width'])
```

```python
X.head()
```

```python
##dependent features
y=iris['target']
```

```python
##train test plot
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
## apply Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
##Visualize the Decision Tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

# Post pruning

### since we constructed decision tree to completely to its depth it has undergone overfitting,we will now overcome overfitting by post pruning the tree (constructing whole tree then pruning to ceartain depth to avoid overfitting),please observe the image very carefully ,now im pruning the above constructed tree to deapth 2 because if you carefully observe ,at that depth one category is completely dominating other two categories ,then what is the reason to split from there......

```python
post_pruning_tree=DecisionTreeClassifier(max_depth=2)
post_pruning_tree.fit(X_train,y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(post_pruning_tree,filled=True)
y_pred=post_pruning_tree.predict(X_test)
```

```python
# calculating metrics for post pruning tree
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
