### **Regression Types**

Regression is about **predicting a number** based on some input. Think of it like drawing a line (or curve) that best fits your data points.

1. **Simple Regression**

   * Only **one input variable** predicts one output.
   * Example: Predict a student’s score based on **hours studied**.
   * Visual: A straight line through a scatter of points.

2. **Multiple Regression**

   * Uses **more than one input variable** to predict the output.
   * Example: Predict score based on **hours studied** and **hours slept**.
   * Visual: A plane or hyperplane through data points in higher dimensions.

3. **Polynomial Regression**

   * The relationship between input and output is **curvy**, not straight.
   * Example: Predict growth of a plant where growth speeds up, then slows down.
   * Visual: A curve that bends to fit the data instead of a straight line.

---

### **Underfitting vs Overfitting**

1. **Underfitting**

   * Model is too **simple** to capture the trend.
   * Result: Bad predictions both on old and new data.
   * Example: Fitting a straight line to a very wiggly dataset.

2. **Overfitting**

   * Model is too **complex** and memorizes the training data.
   * Result: Works perfectly on old data but poorly on new data.
   * Example: Drawing a crazy squiggly curve that passes through every training point.

**Goal:** Find the **balance** – complex enough to learn patterns but simple enough to generalize.

---

### **Error Metrics**

These tell us **how good our model is**:

1. **MAE (Mean Absolute Error)**

   * Average of the **absolute differences** between predicted and actual values.
   * Easy to understand: “On average, our prediction is off by X units.”

2. **MSE (Mean Squared Error)**

   * Average of **squared differences**.
   * Penalizes **big mistakes more** than small ones.

3. **RMSE (Root Mean Squared Error)**

   * Square root of MSE.
   * Back to the **same units** as the original data, easy to interpret.

---

### **R² (R-squared) & Adjusted R²**

1. **R²**

   * Shows **how much of the variation** in the data your model explains.
   * 1 = perfect fit, 0 = model explains nothing.

2. **Adjusted R²**

   * Like R², but **penalizes unnecessary variables**.
   * Useful in multiple regression to check if extra inputs really help.

---

✅ **Summary in One Sentence:**

* Simple/Multiple/Polynomial Regression: How we draw a line/curve through data.
* Underfitting/Overfitting: Too simple vs too complex.
* MAE/MSE/RMSE: How wrong our predictions are.
* R²/Adjusted R²: How well our model explains the data.
